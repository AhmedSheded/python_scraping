```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.github.com/sheded'
respose = requests.get(url)
soup = BeautifulSoup(respose.content, 'lxml')
```

```python
imags_tags = soup.find_all('img')
len(imags_tags)
```

```python
for i, img in enumerate(imags_tags):
    src = img.get('src')
    # if src.startswith('http'):
    if src[:4] == 'http':
        image = requests.get(src)
        filname = f'images/img_{i}_.jpg'
        with open(filname, 'wb') as f:
            f.write(image.content)
        print(f'Downloaded: {filname}')
```
